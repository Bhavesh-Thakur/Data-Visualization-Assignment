# 5.  Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate a synthetic dataset with correlated features
np.random.seed(0)
n_samples = 100
X = np.random.rand(n_samples, 3)  # Generate random features
# Create correlations
X[:, 1] = 0.5 * X[:, 0] + np.random.normal(0, 0.1, n_samples)  # Correlated with X[:, 0]
X[:, 2] = 0.2 * X[:, 0] + 0.8 * X[:, 1] + np.random.normal(0, 0.1, n_samples)  # Correlated with X[:, 0] and X[:, 1]

# Convert to DataFrame for easier manipulation
df = pd.DataFrame(X, columns=['Feature 1', 'Feature 2', 'Feature 3'])

# Calculate the correlation matrix
corr_matrix = df.corr()

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, center=0)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate the data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Create a DataFrame
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Plotting the 3D surface plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Create the surface plot
surf = ax.plot_surface(x, y, z, cmap='viridis', edgecolor='none')

# Adding titles and labels
ax.set_title('3D Surface Plot of z = sin(sqrt(x^2 + y^2))')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

# Add a color bar which maps values to colors
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=5)

# Show plot
plt.show()
```

# 2. Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns

```python
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Prepare the output notebook
output_notebook()

#  Generate data
np.random.seed(42)  # For reproducibility
n_points = 100
x = np.random.uniform(0, 10, n_points)
y = np.random.uniform(0, 10, n_points)
sizes = np.random.uniform(10, 50, n_points)  # Sizes of markers
colors = np.random.choice(['red', 'blue', 'green', 'orange', 'purple'], n_points)  # Colors of markers

# Create a DataFrame
data = pd.DataFrame({'x': x, 'y': y, 'sizes': sizes, 'colors': colors})

#  Create a Bokeh figure
p = figure(title="Scatter Plot with Variable Sizes and Colors", x_axis_label='x', y_axis_label='y', width=800, height=400)

#  Add scatter plot
p.scatter('x', 'y', size='sizes', color='colors', source=data, legend_field='colors', fill_alpha=0.6)


show(p)
```

# 3. . Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
 fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']

 counts = [20, 25, 30, 35]

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import factor_cmap
from bokeh.models import ColumnDataSource

# Prepare the output notebook
output_notebook()

# Create the data
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

#  Create a Bokeh figure
p = figure(x_range=fruits, title="Fruit Counts", x_axis_label='Fruit', y_axis_label='Count', width=800, height=400)

# Add the bar chart
p.vbar(x='fruits', top='counts', width=0.9, source=source, legend_field="fruits",
       line_color='white', fill_color=factor_cmap('fruits', palette=['#c9d9d3', '#718dbf', '#e84d60', '#ddb7b1'], factors=fruits))

#  Customize the plot
p.legend.orientation = "horizontal"
p.legend.location = "top_center"

show(p)
```

# 4. Create a Bokeh histogram to visualize the distribution of the given data.

 data_hist = np.random.randn(1000)

 hist, edges = np.histogram(data_hist, bins=30)

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

# Prepare the output notebook
output_notebook()

# Generate the data and compute the histogram
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

# Step 3: Create a ColumnDataSource
source = ColumnDataSource(data=dict(top=hist, left=edges[:-1], right=edges[1:]))

# Step 4: Create a Bokeh figure
p = figure(title="Histogram of Data", x_axis_label='Value', y_axis_label='Frequency', width=800, height=400)

# Step 5: Add the histogram to the plot
p.quad(top='top', bottom=0, left='left', right='right', source=source, fill_color='navy', line_color='white', alpha=0.6)

show(p)
```

# 5. Create a Bokeh heatmap using the provided dataset.

 data_heatmap = np.random.rand(10, 10)

 x = np.linspace(0, 1, 10)

 y = np.linspace(0, 1, 10)

 xx, yy = np.meshgrid(x, y)

```python
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import linear_cmap
from bokeh.models import ColorBar, ColumnDataSource
from bokeh.layouts import column
from bokeh.palettes import Viridis256

# Step 1: Prepare the output notebook
output_notebook()

# Step 2: Generate the data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Flatten the data for Bokeh
data = {
    'x': xx.flatten(),
    'y': yy.flatten(),
    'value': data_heatmap.flatten()
}
source = ColumnDataSource(data)

# Step 3: Create a Bokeh figure
p = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y', tools="hover", tooltips=[("x", "@x"), ("y", "@y"), ("value", "@value")], width=800, height=800)

# Step 4: Add the heatmap
mapper = linear_cmap(field_name='value', palette=Viridis256, low=min(data['value']), high=max(data['value']))

p.rect(x='x', y='y', width=1/10, height=1/10, source=source, line_color=None, fill_color=mapper)

# Step 5: Add a color bar
color_bar = ColorBar(color_mapper=mapper['transform'], width=8, location=(0,0))
p.add_layout(color_bar, 'right')

# Step 6: Show the results
show(p)
```
